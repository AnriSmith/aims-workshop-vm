## Clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.cm as cm
```

```python
X = df_continuous.values
reps = 5
```

```python
within_cluster = []
silhouette = [] 
x_range = range(2,41)
for k in x_range:
    temp_wc = []
    temp_si = []
    for i in range(reps):
        clf = KMeans(n_clusters=k,n_jobs=-1)
        clf.fit(X)
        temp_wc.append(clf.inertia_)
        temp_si.append(silhouette_score(X, clf.predict(X)))
    within_cluster.append(temp_wc)
    silhouette.append(temp_si)
within_cluster = np.array(within_cluster)
silhouette = np.array(silhouette)
```

```python
pyplot.errorbar(x_range,np.mean(within_cluster,axis=1),
                yerr = np.std(within_cluster,axis=1), linewidth=linewidth)
plt.ylabel("Within cluster sum of squares")
plt.xlabel("$k$")
plt.grid()
```

```python
pyplot.errorbar(x_range,np.mean(silhouette,axis=1),
                yerr = np.std(silhouette,axis=1), linewidth=linewidth)
plt.ylabel("Average Silhouette Coefficient")
plt.xlabel("$k$")
plt.grid()
```

```python
within_cluster = []
x_range = range(2,41)
for k in x_range:
    temp_wc = []
    for i in range(reps):
        clf = KMeans(n_clusters=k,n_jobs=-1)
        clf.fit(X_standardised)
        temp_wc.append(clf.inertia_)
    within_cluster.append(temp_wc)
within_cluster = np.array(within_cluster)
```

```python
k=7
clf = KMeans(n_clusters=k,n_jobs=-1)
clf.fit(X_standardised)
cluster_labels = clf.predict(X_standardised)
```

```python
x_in = -3
y_in = 1
plt.scatter(X_standardised[:,x_in],
            X_standardised[:,y_in], c=cluster_labels,
            s=70,label=cluster_labels)
pyplot.scatter(clf.cluster_centers_[:,x_in],clf.cluster_centers_[:,y_in],marker='x',s=150,c=range(k))
plt.title("K-Means Clustering $k$=" + str(k)+ " Scatter: Storage Size (GB) vs Original Buying Price")
plt.xlabel("Storage Size (GB)")
plt.ylabel("Original Buying Price")
plt.grid()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca.fit(X_standardised)
```

```python
X_pca = pca.transform(X_standardised)
X_pca.shape
```

```python
plt.scatter(X_pca[:,0],X_pca[:,1], c=cluster_labels,
            s=70,label=cluster_labels)
pca_centers = pca.transform(clf.cluster_centers_)
pyplot.scatter(pca_centers[:,0],pca_centers[:,1],marker='x',s=150,c=range(k))
plt.title("K-Means Clustering $k$=" + str(k)+ " Scatter: Storage Size (GB) vs Original Buying Price")
plt.xlabel("PC 1")
plt.ylabel("PC0")
plt.grid()
```
